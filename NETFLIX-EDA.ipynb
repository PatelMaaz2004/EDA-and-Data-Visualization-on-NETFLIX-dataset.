## Movies V/S TV Shows 

```python
sns.countplot(data=netflix, x='type', palette='rainbow')
plt.title('Number of Movies vs TV Shows on Netflix')
plt.xlabel('Type of Content')
plt.ylabel('Count')
plt.show()
```

### Insight: Here we can see that number of movies are more than TV shows on Netflix.

----

## Content adding per year on netflix

```python
content_per_year = netflix['year_added'].value_counts().sort_index()
content_per_year.plot(kind='bar', color='skyblue')
plt.title('Content Added to Netflix Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()  
```

### Insight: Here we can see that Netflix started uploading content from 2008 and it started growing from year to year and the maximum content was uploaded in the year 2019 .

## How much content is uploaded on Netflix.

```python
netflix['year_added'].value_counts().sum()  
# This is also an insight, Netflix has uploaded 8797 contents till the year 2021.
```

## Top countries producing most contents.

```python
top_countries = netflix['country'].value_counts().head(10)
sns.barplot(x=top_countries.values,y=top_countries.index,palette='coolwarm')
plt.title('Top 10 Countries with Most Netflix Titles')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

### Insight: The country producing the most content on Netflix is United States.

---

## Most common ratings

```python
sns.countplot(data=netflix, x='rating', order=netflix['rating'].value_counts().index, palette='rainbow')
plt.title('Distribution of Netflix Ratings')
plt.xticks(rotation=45)
plt.show()
```

### Insight: Most of the Netflix content is rated TV-MA and PGA-13.Focusing on adult contents and teen audiences.

---

## Most popular Genres.

```python
netflix['listed_in'] = netflix['listed_in'].fillna('Unknown')
genres = netflix['listed_in'].str.split(',').explode().str.strip()
top_genres = genres.value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='cubehelix')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

### Insight: Drama, International Movies, and Comedians are the most popular genres on Netflix.

## Month wise addition of content on Netflix.

```python
netflix['month_added'].value_counts().sort_index()
```

```python
netflix['month_added'] = netflix['date_added'].dt.month

month_counts = netflix['month_added'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=month_counts.index, y=month_counts.values, palette='plasma')
plt.title('Netflix Titles Added per Month')
plt.xlabel('Month')
plt.ylabel('Number of Titles')
plt.show()
```

### Insight: This graph shows in which months Netflix adds more titles. As we can see that Netflix adds the most content in July and the least in February, which suggests that they release more shows during mid-year, probably when more people are on vacation.
